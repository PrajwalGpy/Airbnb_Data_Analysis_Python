### **Task 4 : Data Cleaning**

```python
data.isnull().sum()

# removing all missing values
data.dropna(inplace=True)

data.isnull().sum()
```

```python
data.shape
```

```python
# dealing with duplicates rows
data.duplicated().sum()

# show the duplicates
data[data.duplicated()]

# deleting all duplicates rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# type casting
# Changing data types
data.dtypes

data['id'] = data['id'].astype(object)
data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

#### **Task 5 : Data Analyses**

##### Uniovariant Analyses

```python
# identifying outliers in price

df =data[data['price'] < 1500]

sns.boxplot(data=df,x='price')
```

```python
# Price Distribution

plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# availability_365 Distribution

plt.figure(figsize=(6,4))
sns.histplot(data=df,x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# AVG price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean().round(2)
```

##### **Bi Variable Analyses**

```python
# price dependency on neighbourhood_group
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.show()
```

```python
# number of  reviews and price rel

plt.Figure(figsize=(8,5))
plt.title('Locality and Review dependency')
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df,vars=[ 'price','minimum_nights', 'number_of_reviews', 'availability_365'],hue='room_type')
plt.show()
```

```python
# Heat Map - Correlation of  one variable with others for numerical column

corr = df[['latitude','longitude','price','minimum_nights','number_of_reviews','availability_365','beds']].corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data=corr,annot=True)
plt.show()
```
